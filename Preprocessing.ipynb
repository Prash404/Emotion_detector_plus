```python
import os
import shutil


source_directory = 'C:/Users/prash/anaconda3/Internwork/Project1/UTKFace'
target_directory = 'C:/Users/prash/anaconda3/Internwork/Project1/UTKAge'

if not os.path.exists(source_directory):
    print("The specified source directory does not exist")
    exit()

if not os.path.exists(target_directory):
    os.makedirs(target_directory)

age_ranges = [(1, 6), (7, 12), (13, 17), (18, 29), (30, 59), (60, 120)]

def get_age_range_directory(age):
    for start, end in age_ranges:
        if start <= age <= end:
            return f"{start}-{end}"
    return None

for filename in os.listdir(source_directory):
    if filename.endswith(".jpg"):
        parts = filename.split('_')
        age = int(parts[0])
        
        age_range = get_age_range_directory(age)
        
        if age_range:
            age_directory = os.path.join(target_directory, age_range)
            if not os.path.exists(age_directory):
                os.makedirs(age_directory)
            
            source_path = os.path.join(source_directory, filename)
            destination_path = os.path.join(age_directory, filename)
            shutil.copy(source_path, destination_path)
            print(f"Copied {filename} to {age_directory}")

print("Images have been successfully copied and sorted by age ranges in the UKTage folder.")
```

```python
import os
import matplotlib.pyplot as plt

target_directory = 'C:/Users/prash/anaconda3/Internwork/Project1/UTKA/Train'
age_ranges = [(1, 6), (7, 12), (13, 17), (18, 29), (30, 59), (60, 120)]

age_distribution = {f"{start}-{end}": 0 for start, end in age_ranges}

for age_range in age_distribution.keys():
    age_range_directory = os.path.join(target_directory, age_range)
    if os.path.exists(age_range_directory):
        age_distribution[age_range] = len(os.listdir(age_range_directory))

for age_range, count in age_distribution.items():
    print(f"Age range {age_range}: {count} images")

age_ranges = list(age_distribution.keys())
counts = list(age_distribution.values())

plt.figure(figsize=(10, 6))
plt.bar(age_ranges, counts, color='skyblue')
plt.xlabel('Age Ranges')
plt.ylabel('Number of Images')
plt.title('Distribution of Images Across Age Ranges')
plt.xticks(rotation=45)
plt.show()
```

```python
import os
import shutil
import random
import cv2

source_directory = 'C:/Users/prash/anaconda3/Internwork/Project1/UTKAge'
train_directory = 'C:/Users/prash/anaconda3/Internwork/Project1/UTKA/Train'
test_directory = 'C:/Users/prash/anaconda3/Internwork/Project1/UTKA/Test'

if not os.path.exists(source_directory):
    print("The specified source directory does not exist")
    exit()

if not os.path.exists(train_directory):
    os.makedirs(train_directory)
if not os.path.exists(test_directory):
    os.makedirs(test_directory)

for age_range in os.listdir(source_directory):
    age_range_path = os.path.join(source_directory, age_range)

    if os.path.isdir(age_range_path):

        all_files = [f for f in os.listdir(age_range_path) if f.endswith(".jpg")]

        random.shuffle(all_files)

        split_point = int(len(all_files) * 0.7)

        train_files = all_files[:split_point]
        test_files = all_files[split_point:]

        def copy_files(file_list, target_subdir):
            target_subdir_path = os.path.join(target_subdir, age_range)
            if not os.path.exists(target_subdir_path):
                os.makedirs(target_subdir_path)
            for filename in file_list:
                source_path = os.path.join(age_range_path, filename)
                destination_path = os.path.join(target_subdir_path, filename)
                image = cv2.imread(source_path)
                gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                cv2.imwrite(destination_path, gray_image)
                print(f"Copied {filename} to {target_subdir_path}")

        copy_files(train_files, train_directory)
        copy_files(test_files, test_directory)

print("Files have been successfully divided")
```
