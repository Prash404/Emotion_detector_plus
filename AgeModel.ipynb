```python
import matplotlib.pyplot as plt
from glob import glob
import cv2
import random
import os
%matplotlib inline

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, Input, Dropout, Flatten, Conv2D
from tensorflow.keras.layers import BatchNormalization, Activation, MaxPooling2D, GlobalAveragePooling2D, AveragePooling2D
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.utils import plot_model

from IPython.display import SVG, Image
import tensorflow as tf
```

```python
def Convolution(input_tensor, filters, kernel_size, pool_type='max'):
    x = Conv2D(filters=filters, kernel_size=kernel_size, padding="same")(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    if pool_type == 'max':
        x = MaxPooling2D(pool_size=(2, 2))(x)
    else:
        x = AveragePooling2D(pool_size=(2, 2))(x)
    x = Dropout(0.4)(x)
    return x
```

```python
def Dense_f(input_tensor, nodes):
    x = Dense(nodes)(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(0.5)(x)
    return x
```

```python
def model_f(input_shape):
    inputs = Input(input_shape)

    conv_1 = Convolution(inputs, 64, (3, 3), 'max')
    conv_2 = Convolution(conv_1, 128, (5, 5), 'max')
    conv_3 = Convolution(conv_2, 128, (3,3), 'max')
    
    gap = Flatten()(conv_3)
    dense_0 = Dense_f(gap, 256)
    dense_1 = Dense_f(dense_0, 512)
    
    output = Dense(6, activation="softmax")(dense_1)
    model = Model(inputs=inputs, outputs=output)

    optimizer = Adam(learning_rate=0.0005)
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

    return model
```

```python
model = model_f((64,64,1))
model.summary()
```

```python
image_size = 64
batch_size = 64

datagen_train = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    zoom_range=0.1, 
    horizontal_flip=True 
)

train_generator = datagen_train.flow_from_directory(
    "UTKA/Train/",
    target_size=(image_size, image_size),
    color_mode="grayscale",
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=True
)

datagen_validation = ImageDataGenerator(rescale=1./255)

validation_generator = datagen_validation.flow_from_directory(
    "UTKA/Test/",
    target_size=(image_size, image_size),
    color_mode="grayscale",
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=False
)
```

```python
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=0.00001, verbose=1)
early_stopping = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True, verbose=1)

epochs = 20
steps_per_epoch = train_generator.n//train_generator.batch_size
validation_steps = validation_generator.n//validation_generator.batch_size

checkpoint = ModelCheckpoint("age_model_weights1.h5", monitor='val_accuracy', save_weights_only = True, mode = 'max', verbose = 1)
callbacks = [checkpoint, early_stopping, reduce_lr]
```

```python
history = model.fit(
    x = train_generator, 
    steps_per_epoch = steps_per_epoch, 
    epochs = epochs, 
    validation_data = validation_generator, 
    validation_steps = validation_steps, 
    callbacks = callbacks
)
```

```python
model.evaluate(validation_generator)
```

```python
model_json = model.to_json()
with open("model_age1.json","w") as json_file:
    json_file.write(model_json)
```
